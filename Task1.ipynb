```python
for i in df.select_dtypes(np.number):
    sns.boxplot(df[i],color='purple')
    plt.axhline(df[i].mean(),color='b')
    plt.axhline(df[i].median(),color='g')
    print('mean :',df[i].mean())
    print('median :',df[i].median())
    print('mode :',df[i].mode())
    print('skew :',round(df[i].skew(),2))
    print('Kurtosis :',round(df[i].kurt(),2))
    print('std :',df[i].std())
    print('var :',df[i].var())
    print('coeff_of_variation :',df[i].std()/abs(df[i].mean()))
    plt.title(i,color='r')
    plt.show()
```

```python
x=df.drop(['Species'],axis=1)
y=df['Species']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.3,random_state=70)
```

```python
print('Training Data :',xtrain.shape,ytrain.shape)
print('Testing Data :', xtest.shape,ytest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x,y)
```

```python
expected = y
pred=lr.predict(x)
```

```python
from sklearn import metrics
print(metrics.classification_report(expected, pred))
```

```python
print(metrics.confusion_matrix(expected, pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(expected, pred)
```

```python
new = np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
prediction = lr.predict(new)
print("Prediction of Species: {}".format(prediction))
```
